```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movie3=pd.read_csv('movie2.csv')
```

```python
sns.boxplot(x='RunTime',data=movie3)
plt.show()
```

```python
Q1 = movie3['RunTime'].quantile(0.25)
Q3 = movie3['RunTime'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
runtime_outliers= movie3[(movie3['RunTime'] < lower_bound) | (movie3['RunTime'] > upper_bound)]
runtime_outliers
```

```python
observations = len(movie3)
total_runtime_outliers = len(runtime_outliers)
percentage_runtime_outliers = (total_runtime_outliers / observations) * 100
percentage_runtime_outliers
```

```python
#the runtime of the outliers match and
#the outliers are merely 1% so, we will leave them and move on witn the analysis
```

```python
runtime_median=movie3['RunTime'].median()
runtime_median
```

```python
movie3['RunTime'].fillna(runtime_median,inplace=True)
```

```python
movie3.isnull().sum()
```

```python
movie3.head()
```

```python
#renaming for better understanding
cast = movie3.rename(columns={'STARS':'CREW'})
movie3 = cast 
movie3
```

```python
description = movie3.rename(columns={'ONE-LINE':'DESCRIPTION'})
movie3 = description
movie3
```

```python
#the \n in genre, description & cast are really distracting lets deal with it 
movie3['DESCRIPTION'] = movie3['DESCRIPTION'].str.strip()
movie3['GENRE'] = movie3['GENRE'].str.strip()
movie3['CREW'] = movie3['CREW'].str.strip()
```

```python
movie3.head()
```

```python
#remove \n in betn of cast
movie3['CREW'] = movie3['CREW'].str.replace('\n', ' ')
```

```python
movie3.head()
```

```python
movie3['CREW']
#the crew column is looking messy lets split it into 2 columns for better understanding
```

```python
Final_movie = movie3.join(movie3['CREW'].str.partition('|').rename({0: 'DIRECTORS',1:'TY', 2: 'CAST'}, axis=1))
Final_movie
```

```python
#dropping columns we do not need 
Final_movie.drop('TY',axis = 1,inplace = True)
Final_movie.drop('CREW',axis = 1,inplace = True)
```
