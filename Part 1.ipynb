```python
duplicate_mask = movie.duplicated(keep=False)
duplicate_rows = movie[duplicate_mask]
duplicate_rows
```

```python
plt.figure(figsize=(10,6))
movie['YEAR'].hist(bins=10,edgecolor='k')
plt.show()
```

```python
movie['YEAR']
#we can see the date format is diff could be input error 
#need to handle these and turn into correct format
```

```python
def clean_year(date_str):
    if pd.isna(date_str) or not isinstance(date_str, str):
        return None
    cleaned_str = date_str.strip('()').replace('–', '').strip()
    if len(cleaned_str) == 4 and cleaned_str.isdigit():
        return int(cleaned_str)
    elif len(cleaned_str) == 9 and cleaned_str[:4].isdigit() and cleaned_str[4] == '-' and cleaned_str[5:].isdigit():
        return cleaned_str[:4], cleaned_str[5:]
    return None
cleaned_years = movie['YEAR'].apply(clean_year)
print(cleaned_years)
```

```python
movie['YEAR']=cleaned_years
movie['YEAR']
```

```python
plt.figure(figsize=(10,6))
movie['YEAR'].hist(bins=10,edgecolor='k')
plt.show()
```

```python
movie['YEAR'].isnull().sum()
#the null value has changed lets check for outliers
```

```python
Q1 = movie['YEAR'].quantile(0.25)
Q3 = movie['YEAR'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = movie[(movie['YEAR'] < lower_bound) | (movie['YEAR'] > upper_bound)]
outliers
#the authencity of the dates are correct so we will keep it as it is
```
